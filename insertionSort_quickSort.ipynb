```python
# 단순 삽입법
# 1. 기준을 잡는다 기준은 우선 정렬됬다고 가정한다
# 2. 다음 index를 정렬영역으로 들여온다.
# 3. 그 영역 안에서 자신의 자리를 찾아 들어간다.(정렬)
 
# for i=1 부터 arr마지막까지{
#     for j=i 부터 0까지 do{
#         if arr[j-1] > arr[j]{
#             arr[j-1], arr[j] = arr[j], arr[j-1];
#         }
#     }
# }

def sort(arr):
    for i in range(1,len(arr)): 
        #i는 1부터(0번index는 이미 정렬된 상태라고 가정한다.)
        #i번 index는 우선 정렬영역으로 들어오고 아래에서 정렬한다.
        for j in range(i,0,-1): #j는 i부터 0까지(정렬영역)
            if arr[j-1] > arr[j]: #j와 j-1을 비교한다.
                arr[j-1], arr[j] = arr[j], arr[j-1]#조건에 따라 교환한다.
                #반복을통해 정렬영역에 들어온 모든값을 제자리에 둔다.
            else:
                break
                #arr[j-1] > arr[j]이 거짓이라면 더이상 index를 
                #검사할 필요가 없으므로(이미 정렬 상태이기때문) 
                #무의미한 반복을 줄이기 위해 break로 탈출
    return arr

arr = [48, 9, 36, 21, 53, 1, 12, 3, 25]
print(sort(arr))
```

```python
# 퀵정렬
# 1.기준을 잡는다.
# 2.기준의 앞/뒤로 작은 값과 큰 값을 나누어 배치한다.(앞 / 뒤로 집합이 생김)
# 3.나누어진 data들의 집합에서 1~2반복한다. 
# (반복하다가 집합의 크기가 1이하면 종료)

# 이전의 정렬방식과 다른점은 정렬되어 탐색범위에서 제외되는 index가 배열의 한쪽 끝이 아닌 
# 중앙에 위치할 수도 있다
# 기준을 잡고 앞/뒤를 다른 집합으로보고 그 내부에서 다시 퀵정렬 반복
# 앞/뒤를 다른 집합으로나누는건 아래와 같이 해봄

# 1. 앞 / 뒤를 별도의 배열로 만든다.
# (기준을 잡을 때마다 집합이 나뉘므로 탐색해야할 범위 즉,새로운 배열이 저장될 것이다.)

# 2. 앞 / 뒤의 시작index와 끝index를 별도로 저장하여 퀵정렬 알고리즘을 반복시킨다.
# (기준을 잡을 때마다 집합이 나뉘므로 탐색해야할 범위 즉, 시작과 끝 index가 바뀔것이다.)
```

```python
# 배열 이용하여 구현
# 0.정렬할 배열을 arr, 기준으로 나뉜 앞의 집단은 arr_f, 뒤의 집단은 arr_r, 
#   기준은 pivot변수 사용
# 1.정렬할 배열을 입력받는다.
# 2.배열에서 기준을 잡는다.(여기서는 가장 앞 index를 기준으로 잡는다.)
# 	2-1. pivot에 해당 배열의 가장 앞 index를 저장한다.
# 3.기준의 앞에는 작은 데이터/기준의 뒤에는 큰 데이터
# 	3-1. pivot과 배열의 데이터를 비교한다.
# 	3-2. pivot보다 작으면 arr_f에 저장한다. pivot보다 크면 arr_r에 저장한다.
# 	3-4. 현재 arr을 순서대로 보면 arr_f + arr[pivot] + arr_r
# 4. 반복의 탈출은 arr의 길이가 1이하일때로 생각하는데 반복문으로는 힘들거 같다.
#    재귀함수를 사용하되 return에서 연산은 피한다.

# def sort(arr)
#     if arr의 길이가 1이하면{
#         arr 반환
#     }
#     for i=1부터 arr의 마지막까지{
#         if arr[pivot] < arr[i]{
#             arr_f에 arr[i] 추가;
#         }else{
#             arr_r에 arr[i] 추가;
#         }
#     }
#     (result = arr_f + arr[pivot] + arr_r)이부분이 재귀로 들어가야 좋을듯함
#     result = sort(arr_f) + 여기에 arr[pivot] + sort(arr_r)
#     실제코드에서는 불가능하고 다른방법써야함
#     return result
# end sort()

def sort(arr):
    if len(arr) <= 1: 
        #여기서 len(arr)==1이 안되는 이유는 남은 배열의 집합이 모두 기준보다 크거나 작으면
        #한쪽 배열만 값이 들어가고 나머지는 값이 비어서 len(arr)이 0으로 될 수도 있기때문
        return arr
    pivot = 0  #기준 index 항상 배열의 첫번째 index를 가리킴
    arr_f = [] #앞쪽 집합
    arr_r = [] #뒤쪽 집합
    
    for i in range(1,len(arr)): #pivot(기준)이 0이기때문에 i는 1부터 시작
        if arr[pivot] > arr[i]:
            arr_f.append(arr[i])
        else:
            arr_r.append(arr[i])
    
    arr_f.append(arr[pivot]) #배열의 순서는 arr_f 다음 arr[pivot] 다음 arr_r 이 와야한다.
    result = sort(arr_f) + sort(arr_r) 
    #return에서 연산을 피하기위해 result변수 사용
    return result
            
arr = [20, 36, 21, 53, 1, 12, 45, 3, 25]
print(sort(arr))
```

```python
#인덱스 활용하여 구현  in place

##중복값 해결완료

# 0.정렬할 배열을 arr, 시작 index는 begin, 끝 index는 end, 기준은 pivot
#   0-1.비교할때 사용할 인덱스변수 f, r 만든다
# 1.정렬할 배열을 입력받는다.
# 2.배열에서 임의로 첫 기준을 잡는다.(여기서는 중앙 index를 기준으로 잡는다.)
# 	2-1. pivot에 해당 배열의 평균으로 중앙 index를 설정한다.
# 3.기준의 앞과 뒤에서 각각 조건에 맞는 값을 탐색한다.
# 	3-1. pivot을 기준으로 비교한다.
# 	3-2. f변수를 배열의 시작 index부터 +1씩 이동하며 pivot보다 *크거나 같은값을 찾는다.
# 	3-3. r변수를 배열의 끝 index부터 -1씩 이동하며 pivot보다 *작은 값을 찾는다.
# 4.결과에 따른 값의 교환
#   4-1. f index < r index(교차되지 않은 상태)
#            둘 다 기준에 도착하기 전에 만족하는 값을 찾았으므로 arr[f]과 arr[r]의 위치를 교환
#   4-2. f >= r
#            1)f와 r이 같은곳에서 멈추면 pivot과 현재 index를 교환 
#               ***교환해도 지금 위치부터*** 다시 탐색
#            2)f와 r이 교차되면 반복을 중지하고 r위치와 pivot위치의 값을 교환
# 5.pivot의 위치가 바뀌면 그 위치의 값은 고정되고 기준이되어 앞뒤가 부분집합으로 나뉜다.
# 6.부분집합의 크기가 1이하가 될 때까지 반복한다.

#     ex) 60 10 30 2 11 7 22
#       f는 0번 인덱스(값 60)/ r은 6번 인덱스(값 22) / pivot은 3번 인덱스(값 2)
#       f는 2보다 크거나 같은값을 찾아가는데 0번이 60이므로 stop
#       r은 2보다 작은값을 찾아가는데 없으므로 계속 감소하다가 60에서 f와 만나게됨
#       f와 r이 만나면 pivot과 교환을 함
#       [2] 10 30 60 11 7 22
#       이제 2는 고정되고 나머지 영역에서 pivot을 재설정함


def mk_piv(arr,begin,end): # 정렬할배열, 시작인덱스, 끝인덱스
    pivot = (begin+end) // 2 #평균으로 중간 인덱스 설정
    f=begin #f는 배열의 시작부터
    r=end #r은 배열의 끝부터
    
    while f < r: #(교차되지 않았다면)
        #만약 f와 r이 같거나 교차된다면 pivot의 요소를 r요소와 교환해야 하므로
        #while문 밖의 코드 실행
        
        
        while arr[f] < arr[pivot] and f < r:
            #arr[pivot]보다 작으면 실행이므로 크거나 같은 값을 찾으면 중지(3-2)
            #f가 증가되더라도 r보다 커지면 안되므로 and로 조건을 추가
            f += 1
        while arr[r] >= arr[pivot] and f < r:
            #arr[pivot]보다 크거나 같으면 실행이므로 작은 값을 찾으면 중지(3-3)
            #r이 감소되더라도 f보다 작아지면 안되므로 and로 조건을 추가
            r -= 1
        
        if f < r: #(4-1)
            arr[f], arr[r] = arr[r], arr[f]
            if f == pivot: 
                #f의 위치와 pivot의 위치가 같으면 r을 기준으로 잡고 
                #pivot을 새로 설정해야함
                #    -> 반환하여 sort()에서 index 범위를 재설정하므로 pivot의 위치가 변경됨
                
                #여기서 pivot과 f가 같을 수 있는 이유는 
                #f는 arr[pivot]보다 '크거나 같은 값'을 찾기때문 즉, pivot에 도달하면 멈춤
                return r 
                #여기서 r을 반환하는 이유는 r이 arr[pivot]보다 작은 값에만 멈추는게 가능하며
                #작은 값이없으면 탐색범위의 시작지점까지 이동함
                #pivot위치를 만나도 조건에 맞는 값을 찾기위해 계속 움직이며
                #교차되거나 f와 같은곳에서 멈출경우
                #r이 위치한 곳이 현재 pivot의 자리를 정하는데 사용됨
                
    #이곳까지 왔다는건 while f < r 조건이 거짓이여서 빠져나온 경우 (교차되거나 같은경우)
    #f와 r이 같은 곳에서 멈추면 현재 위치와 pivot을 교환한다. (4-2)
    #교차되면 반복을 멈추고 pivot과 r의 위치를 교환한다.
    arr[pivot], arr[f] = arr[f], arr[pivot] 
    return r

def sort(arr, begin, end):
    if begin < end:
        p = mk_piv(arr,begin,end)
        sort(arr, begin,p-1)
        sort(arr, p+1, end)

arr = [20, 3,36, 21, 3, 53, 12, 45, 3, 25]
sort(arr, 0, len(arr)-1)
print(arr)
```
